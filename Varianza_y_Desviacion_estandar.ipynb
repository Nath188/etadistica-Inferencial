# importamos librerias necesarias para analisis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# leemos base de datos a trabajar en ruta de origen

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length','sepal-width', 'petal-lenght', 'petal-width', 'class']
iris = pd.read_csv(url,names = names)
```

```python
iris.head()
```

petalo-sepalo.jpeg

```python
plt.hist(iris['sepal-length'], color ='blue', bins =int(180/5))
```

```python
sns.distplot(iris['sepal-length'],hist= True, kde= True, bins= int(180/5), color ='green')
```
